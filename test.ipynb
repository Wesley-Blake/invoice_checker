```python
from pathlib import Path
import os
import shutil
```

```python
invoices = Path("")
```

```python
dict_invoice = {invoice: 0 for invoice in invoices.iterdir()}
print(dict_invoice)
```

```python
search_dir = ""
for root,_,files in os.walk(search_dir):
    for invoice in dict_invoice.keys():
        for file in files:
            if invoice.name[:-4] in file:
                dict_invoice[invoice] += 1
```

```python
move_invoice = [key for key, value in dict_invoice.items() if value == 0]
```

```python
final_dest = Path("")
```

```python
for invoice in move_invoice:
    dest = final_dest / invoice.name
    if dest.exists():
        raise Exception(dest)
    else:
        shutil.move(invoice, dest)
```
